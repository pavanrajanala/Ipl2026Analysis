```python
print("Matches:", matches.shape)
print("Batting:", batting_stats.shape)
print("Bowling:", bowling_stats.shape)
print("Deliveries:", delivery.shape)
print("Points Table:", points_table.shape)
```

```python
print("Matches:", matches.isnull().sum().sum())
print("Batting:", batting_stats.isnull().sum().sum())
print("Bowling:", bowling_stats.isnull().sum().sum())
print("Deliveries:", delivery.isnull().sum().sum())
print("Points:", points_table.isnull().sum().sum())
```

```python
delivery.isnull().sum().sort_values(ascending=False)
```

```python
delivery.drop(
    columns=['c1','c2','c3','c4','c5',
             'c8','c9','c10','c11','c12'],
    inplace=True
)
```

```python
delivery['extras_detail'] = (
    delivery['extras_detail']
    .fillna('No Extra')
)
```

```python
batting_stats.fillna(0, inplace=True)
```

```python
bowling_stats.isnull().sum()
```

```python
bowling_stats.fillna(0, inplace=True)
```

```python
matches.fillna("Unknown", inplace=True)
```

```python
print("Matches:", matches.isnull().sum().sum())
print("Batting:", batting_stats.isnull().sum().sum())
print("Bowling:", bowling_stats.isnull().sum().sum())
print("Deliveries:", delivery.isnull().sum().sum())
print("Points:", points_table.isnull().sum().sum())
```

```python
matches.to_csv('matches_clean.csv', index=False)
batting_stats.to_csv('batting_clean.csv', index=False)
bowling_stats.to_csv('bowling_clean.csv', index=False)
delivery.to_csv('deliveries_clean.csv', index=False)
points_table.to_csv('points_clean.csv', index=False)
```
